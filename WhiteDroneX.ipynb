```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8m.pt")
```

```python
b = model.predict("D:\Pics\P7.jpg")
```

```python
x = model.predict(["D:\Pics\P1.jpg", "D:\Pics\P2.jpg", "D:\Pics\P3.jpg"])
```

```python
len(b[0].boxes)
```

```python
result = b[0]
box = result.boxes[0]

#Basic syntax for co-ordinates, class id and degree of confidence
raw_co_ord = box.xyxy[0].tolist()
co_ord = [round(x) for x in raw_co_ord]

class_id = result.names[box.cls[0].item()]

#Rounded to 2 decimal places
conf = round(box.conf[0].item(), 2)
```

```python
print(co_ord)
print(class_id)
print(conf)
```
